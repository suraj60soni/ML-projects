MECHINE LEARNING PART

selecting necessary columns for maually

```python
new_df=df[['Company','CPU_company','CPU_freq','GPU_model',"Ram","PrimaryStorage","Inches","Price_euros"]]
new_df.head(5)
```

visualy checking reletionship or pattern in this data 


```python
sns.pairplot(data=new_df,hue='Price_euros')
plt.show
```

labeling the data and converting the data categorical,float value to int

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
new_df['Company']=le.fit_transform(new_df['Company'])
new_df['CPU_company']=le.fit_transform(new_df['CPU_company'])
new_df['CPU_freq']=le.fit_transform(new_df['CPU_freq'])
new_df['GPU_model']=le.fit_transform(new_df['GPU_model'])
new_df['PrimaryStorage']=le.fit_transform(new_df['PrimaryStorage'])
new_df['Price_euros']=le.fit_transform(new_df['Price_euros'])

new_df.head(5)
```

data divided into independent variable or dependent variable

```python
input_data=new_df.drop(columns=["Price_euros"],axis=[1])
output_data=new_df[['Price_euros']]
```

spliting the data into train data or test data  in 4 part

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input_data,output_data,test_size=0.20,random_state=70)
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

testing the best model selection

```python
from sklearn.ensemble import RandomForestRegressor
rd=RandomForestRegressor()
rd.fit(x_train,y_train)
pre_rd=rd.predict(x_test)

rd.score(x_test,y_test),rd.score(x_train,y_train)
```

```python
#using cross_validation method 

from sklearn.model_selection import RandomizedSearchCV
```

Evaluate the model with MES or RMSE

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test,pre_rd)
rmse=np.sqrt(mse)
rmse
```

```python

rd.predict([[10,1,7,42,2,3,14.0]]) # checking the accuracy of price 
```
